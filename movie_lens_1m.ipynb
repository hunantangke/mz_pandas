```python
import pandas as pd
unames = ['userid','gender','age','occupation','zip']
```

```python
users = pd.read_table('ml-1m/users.dat',sep='::',header=None,names=unames)
```

```python
print len(users)
users.head()
```

```python
rating_names = ['userid','movieid','rating','timestamp']
ratings = pd.read_table('ml-1m/ratings.dat',sep='::',header=None,names=rating_names)
movie_names = ['movieid','title','genres']
movies = pd.read_table('ml-1m/movies.dat',sep='::',header=None,names=movie_names)
```

```python
print len(ratings)
ratings.head()
```

```python
print len(movies)
movies.head()
```

```python
data = pd.merge(pd.merge(users,ratings),movies)
```

```python
print len(data)
data.head()
```

```python
data[data.userid==1]
```

```python
data[data.movieid==1193]
```

```python
rating_by_gender = data.pivot_table(values='rating',index='title',columns='gender',aggfunc='mean')
```

```python
rating_by_gender.head(10)
```

```python
rating_by_gender['diff'] = rating_by_gender.F - rating_by_gender.M
rating_by_gender.head(10)
```

```python
rating_by_gender.sort_values(by='diff',ascending=False).head(10)
```

```python
rating_by_title = data.groupby('title').size()
rating_by_title.head(10)
```

```python
rating_by_title.sort_values(ascending=False).head(10)
```

```python
mean_rating = data.pivot_table(values='rating',index='title',aggfunc='mean')
```

```python
mean_rating.sort_values(ascending=False).head(10)
```

```python
top_10_hot = rating_by_title.sort_values(ascending=False).head(10)
top_10_hot
```

```python
mean_rating[top_10_hot.index]
```

```python
top_20_score = mean_rating.sort_values(ascending=False).head(20)
```

```python
rating_by_title[top_20_score.index]
```

```python
hot_movies = rating_by_title[rating_by_title>1000]
```

```python
print len(hot_movies)
hot_movies.head()
```

```python
top_10_good_movies = mean_rating[hot_movies.index].sort_values(ascending=False).head(10)
top_10_good_movies
```
